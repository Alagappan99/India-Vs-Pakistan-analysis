```python
import pandas as pd
df=pd.read_excel("India Vs Pakistan.xlsx")
df.head()
```

```python
df.info()
df.isnull().sum()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
df['year'] = df['year'].astype(int)
df['india_runs'] = pd.to_numeric(df['india_runs'], errors='coerce')
df['pak_runs'] = pd.to_numeric(df['pak_runs'], errors='coerce')

```

```python
df['total_runs'] = df['india_runs'] + df['pak_runs']
df['india_margin'] = df['india_runs'] - df['pak_runs']
df['winner_is_india'] = (df['winner'].str.lower().str.strip() == 'india').astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("IndiaVsPakistan_clean.csv", index=False)
```

```python
print("Total_Maches =",len(df))
print("India_wins =",df['winner'].value_counts().get('India',0))
print("Pak_wins =",df['winner'].value_counts().get('Pakistan',0))
print("Win % of India =",round(df['winner_is_india'].mean()*100,2))
```
